**Data Science Course Final Project:** ***Predict H1N1 and Seasonal Flu Vaccines***


***The Problem***

- For the past 2 years, it is evident that the COVID-19 pandemic has forever changed the life of every person on the planet. Getting vaccinated is the assured method to combat and prevent further spread of infectous diseases.

- My goal was to build a model to predict the likelihood of an individual getting vaccinated using dataset taken from the National 2009 H1N1 Flu Survey conducted between October 2009 and June 2010 in response to the 2009 H1N1 pandemic.

***What factors will affect an individual's likelihood to get vaccinated?***

- It is speculated that a person's education and knowledge level plays the biggest part in their decision making in whether or not to get vaccinated. 

- From the initial EDA, it looks like there is a slight positive correlation (though not strong) of increased H1N1 vaccination rates with higher H1N1 knowledge.



- But with this dataset, I'll dive further into a wider range of variables from disease knowledge, risk opinions, behavorial reactions, and even household size to explore how they affect an individual's probability of receiving a vaccine. 

***The Data***

- The dataset contains both train and test set features. There are 36 feature columns that includes a wide variety of categorical variables (such as race, sex, education, locations) and numerical variables to represent knowledge/concern levels along with opinions on vaccine risks and effectiveness. The increasing numerical values corresponds to stronger feelings/opinions on the survey.

- Missing values will be dropped as replacing them with mean values may skew the data, especially when it comes to a person's opinions and knowledge.



***Approach***

- To effectively train my model, I limited the features to only float data types to remove data with random character strings. 

- Tested with Decision Tree and Random Forest Classifier models.

- Decided to go with Random Forest as it yielded the best results and allowed me to tune hyperparameters such as:
    - n_estimators
    - max_features
    - min_samples_leaf
